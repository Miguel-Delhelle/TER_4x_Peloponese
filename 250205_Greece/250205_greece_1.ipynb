Chaque projection possde ses propres coordonnes, qui sont parfaitement arbitraires. Les donnes gographiques sont enregistres avec les coordonnes projetes. Il faut faire attention au systme de projection.

## Quelques formats logiciels

Le standard historique est la donne ShapeFile, qui est un format robuste. Il s'agit d'un ensemble de 5 fichiers au mme nom, chacun apportant un type d'information  la donne qu'affiche QGIS (quivalent ARGGIS, mais gratuit et aussi performant).
- le dessin vecteur (.shp et .shx)
- les donnes attributaires, relies  chacun des entits vecteurs (.dbf)
- la donne de projection (.prj)
- un DOCTYPE (.qmd)
J'aime ce format, car je peux trs facilement bidouiller mes donnes. Le dbf s'ouvre avec n'importe quel tableur, et contient la table attributaire telle qu'elle est affiche dans QGIS.

L'autre format est bien tir du JSON, le [GeoJSON](https://en.wikipedia.org/wiki/GeoJSON). Il stocke les donne selon des balises, de type technologie web. Ce fichier ne contient pas les informations lies  la projection de la donne, c'est la raison pour laquelle la convention est d'utiliser comme coordonnes la longitude/latitude (donne non projete). **Mais** rien ne vous emche d'utiliser les donnes projetes cartographiques avec la projection " plat" dus sytme grec. Je que j'entends, c'est que vous pouvez peut tre adopter le systme de coordonnes cartographique comme systme de coordonnes de l'chiquier du jeu. Pour ce faire, rien de plus simple: enregistrez la donne gographique  tout sauf  WGS 84(EPSG:4326).
