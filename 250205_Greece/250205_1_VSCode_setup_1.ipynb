I'm picking PiP, mostly because it's popular in Linux. Here is a [link](https://pip.pypa.io/en/stable/installation/#get-pip-py) to installing Pip from Pip. Sadly, I didn't record the process when I did it. Feel free to message me if you're having trouble with that.

````python
import sys
!{sys.executable} -m pip install --upgrade pip
````

We could run the same command through this syntax in VS Code's console.

````python
!{sys.executable} -m pip list
````

The aforewritten command should be typed in a PowerShell window, but one might as well type it inside VSCode. The ``!{sys.executable} -m`` is installing the package only in the environnement that is currently running.

Getting help:

````python
!{sys.executable} -m pip -h
````
